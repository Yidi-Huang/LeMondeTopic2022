```python
import io
import os.path
import re
import xml.etree.ElementTree as ET
```

```python
    with open('sortie.xml','r') as f:
        xml = ET.parse(f)
        docs = []
        for article in xml.findall("//analyse"):
            doc =[]
            for token in article.findall("./token"):
                #if token.attrib['pos'] in ['PRON','NOUN']:
                form = token.attrib['forme']
                lemme = token.attrib['lemme']
                pos = token.attrib['pos']
                doc.append(f"{form}/{lemme}/{pos}")
            if len(doc) > 0:
                docs.append(doc)
                
len(docs)
docs[:10]
```

```python
# Tokenize the documents.
import nltk
from nltk.tokenize import RegexpTokenizer

# Split the documents into tokens.
tokenizer = RegexpTokenizer(r'\w+')
for idx in range(len(docs)):
    docs[idx] = str(docs[idx]).lower()  # Convert to lowercase.
    docs[idx] = tokenizer.tokenize(docs[idx])  # Split into words.

# Remove numbers, but not words that contain numbers.
docs = [[token for token in doc if not token.isnumeric()] for doc in docs]

# Remove words that are only one character.
docs = [[token for token in doc if len(token) > 1] for doc in docs]
```

```python
nltk.download('wordnet')
# Lemmatize the documents.
from nltk.stem.wordnet import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
docs = [[lemmatizer.lemmatize(token) for token in doc] for doc in docs]

docs[:10]
```

```python
import gensim
```

```python

# Compute bigrams.
from gensim.models import Phrases

# Add bigrams and trigrams to docs (only ones that appear 20 times or more).
bigram = Phrases(docs, min_count=20)
for idx in range(len(docs)):
    for token in bigram[docs[idx]]:
        if '_' in token:
            # Token is a bigram, add to document.
            docs[idx].append(token)
```
